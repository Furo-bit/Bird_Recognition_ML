```python
# Import dependencies
from tensorflow import keras

# Set seed
keras.utils.set_random_seed(42)
```

```python
!dir './Dataset/'
DATASET_PATH = './Dataset'
```

```python
train_dataset = keras.utils.image_dataset_from_directory(
    directory = DATASET_PATH + "/Train",
    image_size=(224, 224),
    batch_size=32,
    shuffle=True,
    label_mode='int'
)

test_dataset = keras.utils.image_dataset_from_directory(
    directory = DATASET_PATH + "/Test",
    image_size=(224, 224),
    batch_size=32,
    shuffle=False,
    label_mode='int'
)

x_train, y_train = next(iter(train_dataset))
x_test, y_test = next(iter(test_dataset))

print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

```python
import matplotlib.pyplot as plt

first_image = x_train[0].numpy()
first_label = y_train[0].numpy()

# Plotta la prima immagine
plt.imshow(first_image.astype("uint8"))
plt.title('Label: {}'.format(first_label))
plt.show()
```
